```python
import pandas as pd; import matplotlib.pyplot as plt
print("Liberaries Imported Successfully")
```

```python
df = pd.read_csv("modern_languages_grades.csv"); df.head()
```

```python
df.shape, df.dtypes
```

```python
df.describe(include="all")
```

```python
df.isna().sum()
```

```python
print("Before conversion:", df["Grade"].dtype)
df["Grade"] = pd.to_numeric(df["Grade"], errors="coerce")
print("After conversion:", df["Grade"].dtype)
```

```python
print(df["Grade"])
df = df.dropna(subset=["Grade"])
print(df["Grade"])
```

```python
print(df["Student"])
df["Student"] = df["Student"].str.strip()  # إ
df = df.drop_duplicates(subset=["Student","Subject"])
print(df["Student"])
```

```python
avg_grade = df.groupby('Subject')['Grade'].mean()
max_grade = df.groupby('Subject')['Grade'].max()

print("Average Grade:\n", avg_grade)
print("Max Grade:\n", max_grade)
```

```python
df.dtypes
df.shape
```

```python
avg_grade = df.groupby('Subject')['Grade'].mean()
print(avg_grade)
```

```python
overall_avg = df['Grade'].mean()
print("Overall Average Grade:", overall_avg)
```
